```python
import os
import pickle
from shutil import copyfile
```

```python
def read_pickle(filepath):
    with open(filepath, 'rb') as f:
        return pickle.load(f)
    
def dump_pickle(filepath, obj):
    with open(filepath, 'wb') as f:
        pickle.dump(obj, f)
```

```python
images_folder = 'dataset/satellite'

train_val_test = read_pickle("dataset/train_val_test_split.pickle")
test_ids = train_val_test['test']
```

```python
result_folder = 'dataset/test_satellite_sample'
```

```python
for img_id in test_ids[:50]:
    img_filename = img_id + '.png'
    copyfile(os.path.join(images_folder, img_filename), os.path.join(result_folder, img_filename))
```

```python
osm_objects_path = 'dataset/osm_objects/'
```

```python
def get_all_tags(img_osm_object):
    return [o.tag for o in img_osm_object.osm_objects if o.tag != 'ridge' and o.tag != 'valley' ]
```
